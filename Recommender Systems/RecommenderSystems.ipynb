# Import Libraries:

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Get the Data:

```python
cols = ['user_id','item_id','rating','timestamp']
```

```python
df = pd.read_csv('u.data', sep='\t', names=cols)
```

```python
df.head()
```

```python
titles = pd.read_csv('Movie_Id_Titles')
```

```python
titles.head()
```

```python
df = pd.merge(df,titles,on='item_id')
```

```python
df.head()
```

```python
df
```

# Exploratory Data Analysis:

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['No. of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
sns.jointplot(x='rating', y='No. of ratings', data=ratings, alpha=0.5)
```

```python
df.head()
```

```python
moviemat = df.pivot_table(values='rating', index='user_id', columns='title')
```

```python
moviemat.head()
```

```python
ratings.sort_values('No. of ratings', ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
```

```python
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head(10)
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['No. of ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars=corr_starwars[corr_starwars['No. of ratings']>100].sort_values('Correlation', ascending=False)
```

```python
corr_starwars.head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
```

```python
corr_liarliar.head()
```

```python
corr_liarliar.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_liarliar=corr_liarliar.join(ratings['No. of ratings'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar.head(10)
```

```python
corr_liarliar=corr_liarliar[corr_liarliar['No. of ratings']>100].sort_values('Correlation', ascending=False)
```
